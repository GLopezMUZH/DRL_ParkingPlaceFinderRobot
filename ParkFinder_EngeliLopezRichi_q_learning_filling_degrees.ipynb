## Different parking lot filling degrees
This is an experiment to see the changes in the values of the q table with different filling degrees


```
# small parking lot
utils = Utils()

nr_episodes = 60000
Q = {}
last_frames =[]
filling_degree_values = np.linspace(0.01, 0.9, 10)

for filling_degree_value in filling_degree_values:
  agent, lmp, parking_environment, file_sufix = create_parking_scenario(nr_episodes=nr_episodes, 
                                      uniform_distribution_p_value=filling_degree_value,
                                      nr_parking_slots_per_lane=5,
                                      nr_parking_lanes=4,
                                      parking_lane_depth=2,
                                      debug=False,
                                      draw_graph=True,
                                      save_plot_graph=True
                                      )


  Q, last_frame = doLearning(agent=agent, parking_environment=parking_environment, 
                      Q=Q, lmp=lmp, save_qt=True, save_frames=False,
                      plot_rewards=True)

  last_frames.append(last_frame)
  #utils.print_frame_summary(last_frame)

plot_field_from_last_frames(frames=last_frames, field_name='resulting state',
                              x_range=range(len(filling_degree_values)), x_label='filling degree', title_str='Changes in filling degree')
plot_field_from_last_frames(frames=last_frames, field_name='reward',
                              x_range=range(len(filling_degree_values)), x_label='filling degree', title_str='Changes in filling degree')
```
