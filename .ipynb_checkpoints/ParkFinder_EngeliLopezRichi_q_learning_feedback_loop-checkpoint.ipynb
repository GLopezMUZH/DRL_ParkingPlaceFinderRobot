## Values of Q-table when updated in different parking lot configurations
This is an experiment to see the changes in the values of the q table when the same q table is updated based on different parking lot entities having the same filling degree distribution

```python
utils = Utils()

agent, lmp, parking_environment, file_sufix = create_base_test_scenario(nr_episodes=60000)
Q = {}
Q_10 = {}
for state in agent.stateSpacePlus:
    for action in agent.possibleActions:
        Q[state, action] = 0

save_qt=False
plot_rewards=False
last_frames =[]
iterations_nr = 100

for i in range(iterations_nr):
  agent, lmp, parking_environment, file_sufix = create_base_test_scenario(nr_episodes=60000)

  if(i in [0,9,49,99]):
    save_qt = True
    plot_rewards = True
  elif(i in [33,66]):
    save_qt = False
    plot_rewards = True
  else:
    save_qt = False
    plot_rewards = False


  Q, last_frame = doLearning(agent=agent, parking_environment=parking_environment, 
                      Q=Q, lmp=lmp, save_qt=save_qt, save_frames=False,
                      plot_rewards=plot_rewards)

  last_frames.append(last_frame)
  #utils.print_frame_summary(last_frame)

plot_field_from_last_frames(frames=last_frames, field_name='resulting state',
                              x_range=range(iterations_nr), x_label='tested entities', title_str='Changes in tested entities')
plot_field_from_last_frames(frames=last_frames, field_name='reward',
                              x_range=range(iterations_nr), x_label='tested entities', title_str='Changes in tested entities')
```
