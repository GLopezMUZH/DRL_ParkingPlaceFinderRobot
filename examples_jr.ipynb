The network below showcases the simple agent functionality I included. Simply specify `agent = True` to let an agent perform one run in the created parking lot. A run begins with the agent starting in node 0 and navigating through the network until he finds a vacant parking spot. At the moment the agent will pick the first parking spot available. If the agent does not encounter a vacant parking spot in a step, he will randomly choose a neighboring driveway node on which to continue.
